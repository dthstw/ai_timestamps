```python
class Splitter: 
    
    def __init__(self, dataframe, video_id=None, buffer_seconds=5):
        self.dataframe = dataframe
        self.video_id = video_id 
        self.buffer_seconds = buffer_seconds
            
    def open_file(self): 
        # Filter the dataframe based on video_id
        filtered_row = self.dataframe[self.dataframe['video_id'] == self.video_id]
        
        # Get the captions and timestamps paths
        caption_path = filtered_row["captions"].values[0]
        timestamps_path = filtered_row["timestamps"].values[0]
        try:
            with open(caption_path, 'r') as f:
                captions = f.read()
        except FileNotFoundError:
            print(f"Caption file {captions} not found.")
                  
        try:
            with open(timestamps_path, 'r') as f:
                timestamps = f.read()
        except FileNotFoundError:
            print(f"Target file {timestamps} not found.")
            
        return captions, timestamps
    
    def parse_time(self, timestamp):
        try:
            return datetime.strptime(timestamp, '%H:%M:%S')
        except ValueError:
            return datetime.strptime(timestamp, '%M:%S')
        
    def time_to_seconds(self, time_obj):
        return time_obj.hour * 3600 + time_obj.minute * 60 + time_obj.second
    
    def normalize_and_split_lines(self, text):
        # Replace non-breaking spaces with regular spaces
        text = text.replace('\xa0', ' ')
        # Reduce multiple spaces to a single space
        text = re.sub(r' +', ' ', text)
        # Split the text into lines
        lines = text.split('\n')
        return lines
    
    def reformat_lines(self, lines):
        formatted_lines = []
        current_timestamp = None
        current_text = ""
        
        timestamp_pattern = re.compile(r'\d{2}:\d{2}:\d{2} - \d{2}:\d{2}:\d{2}:')
        
        for line in lines:
            stripped_line = line.strip()
            
            if timestamp_pattern.match(stripped_line):
                if current_timestamp:
                    formatted_lines.append(f'{current_timestamp} {current_text.strip()}\n')
                current_timestamp, current_text = stripped_line, ""
            else:
                current_text += " " + stripped_line

        if current_timestamp and current_text:
            formatted_lines.append(f'{current_timestamp} {current_text.strip()}\n')

        return formatted_lines
 
    def split_captions(self, segment_duration=15*60, overlap= 5*60):
        captions, _ = self.open_file()
        parsed_captions = []
        
        if '\xa0' in captions:
            # Normalize and split the text into lines
            captions = self.normalize_and_split_lines(captions)  
        else:
            captions = captions.split('\n')
            
        captions = self.reformat_lines(captions)    
        caption_regex = re.compile(r'(\d{2}:\d{2}:\d{2}) - (\d{2}:\d{2}:\d{2}): (.+)')
        for line in captions:
            match = caption_regex.match(line)
            if match:
                start_time = self.parse_time(match.group(1))
                end_time = self.parse_time(match.group(2))
                text = match.group(3).replace("\\'", "'").replace('\\"', '"')
                parsed_captions.append((start_time, end_time, text))
        if parsed_captions:
            total_duration = self.time_to_seconds(parsed_captions[-1][1])
        else:
            total_duration = 0
            
        segments = []
        current_start = 0
        
        while current_start < total_duration:
            current_end = min(current_start + segment_duration, total_duration)
            segment = []
            
            for start_time, end_time, text in parsed_captions:
                start_seconds = self.time_to_seconds(start_time)
                end_seconds = self.time_to_seconds(end_time)
                
                if start_seconds < current_end and end_seconds > current_start:
                    segment.append(f"{start_time.strftime('%H:%M:%S')} - {end_time.strftime('%H:%M:%S')}: {text}")
            segments.append('\n'.join(segment))
            print(f"Segment from {str(timedelta(seconds=current_start))} to {str(timedelta(seconds=current_end))}")
            
            if current_end >= total_duration:
                break
        
            current_start += (segment_duration - overlap)
            current_start = max(current_start, min(self.time_to_seconds(start_time) for start_time, end_time, text in parsed_captions if self.time_to_seconds(start_time) >= current_start))
            
        return segments

    def map_targets_to_segments(self):
        segments = self.split_captions()
        _, targets = self.open_file()
        print(targets)
        
        target_regex = re.compile(r'((?:\d{1,2}:)?\d{1,2}:\d{2}) - (.+)')
        parsed_targets = []
        
        for line in targets.split('\n'):
            match = target_regex.match(line)
            if match:
                target_time = self.parse_time(match.group(1))
                target_text = match.group(2)
                parsed_targets.append((target_time, target_text))
        result_segments = []
        for idx, segment in enumerate(segments, start=1):
            segment_lines = segment.split('\n')
            segment_start_time = self.parse_time(segment_lines[0].split(' - ')[0]) - timedelta(seconds=self.buffer_seconds)
            segment_end_time = self.parse_time(segment_lines[-1].split(' - ')[1].split(': ')[0])   
            segment_targets = []
            for target_time, target_text in parsed_targets:

                if segment_start_time <= target_time <= segment_end_time:
                    segment_targets.append(f"{target_time.strftime('%H:%M:%S')} - {target_text}")                   
            result_segments.append([segment_lines, segment_targets])
        return result_segments
```
